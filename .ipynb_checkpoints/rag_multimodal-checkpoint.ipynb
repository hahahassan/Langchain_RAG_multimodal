## 1. use unstructure to parse pdf into text chunks, tables, images

```python
import os
import tensorflow as tf
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import ChatOpenAI
from unstructured.partition.pdf import partition_pdf
```

```python

# Extract elements from PDF
def extract_pdf_elements(path, fname):
    """
    Extract images, tables, and chunk text from a PDF file.
    path: File path, which is used to dump images (.jpg)
    fname: File name
    """
    return partition_pdf(
        filename=path + fname,
        extract_images_in_pdf=True,
        infer_table_structure=True,
        chunking_strategy="by_title",
        max_characters=4000,
        new_after_n_chars=3800,
        combine_text_under_n_chars=2000,
        image_output_dir_path=path,
    )


# Categorize elements by type
def categorize_elements(raw_pdf_elements):
    """
    Categorize extracted elements from a PDF into tables and texts.
    raw_pdf_elements: List of unstructured.documents.elements
    """
    tables = []
    texts = []
    for element in raw_pdf_elements:
        if "unstructured.documents.elements.Table" in str(type(element)):
            tables.append(str(element))
        elif "unstructured.documents.elements.CompositeElement" in str(type(element)):
            texts.append(str(element))
    return texts, tables
```

```python
# File path
fpath = "./"
fname = "AttentionIsAllYouNeed.pdf"

# Get elements
raw_pdf_elements = extract_pdf_elements(fpath, fname)
```

```python
# Get text, tables
texts, tables = categorize_elements(raw_pdf_elements)

# Optional: Enforce a specific token size for texts
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=400, chunk_overlap=50
)
joined_texts = " ".join(texts)
texts_400_token = text_splitter.split_text(joined_texts)
```

## 2.1 Text and table summary

```python
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
import config



# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = config.API_KEY


# Generate summaries of text elements
def generate_text_summaries(texts, tables, summarize_texts=False):
    """
    Summarize text elements
    texts: List of str
    tables: List of str
    summarize_texts: Bool to summarize texts
    """

    # Prompt
    prompt_text = """You are an assistant tasked with summarizing tables and text for retrieval. \
    These summaries will be embedded and used to retrieve the raw text or table elements. \
    Give a concise summary of the table or text that is well optimized for retrieval. Table or text: {element} """


    prompt = ChatPromptTemplate.from_template(prompt_text)

    # Text summary chain
    model = ChatOpenAI(temperature=0, model="gpt-4o-mini")
    

    summarize_chain = {"element": lambda x: x} | prompt | model | StrOutputParser()

    # Initialize empty summaries
    text_summaries = []
    table_summaries = []

    # Apply to text if texts are provided and summarization is requested
    if texts and summarize_texts:
        text_summaries = [summarize_chain.invoke({"element": text}) for text in texts]
    elif texts:
        text_summaries = texts

    # Apply to tables if tables are provided
    if tables:
        table_summaries = [summarize_chain.invoke({"element": table}) for table in tables]

    return text_summaries, table_summaries




text_summaries, table_summaries = generate_text_summaries(
    texts_400_token, tables, summarize_texts=False
)

```

## 2.2 Image summary

```python
import base64
from langchain_core.messages import HumanMessage


def encode_image(image_path):
    """Getting the base64 string"""
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


def image_summarize(img_base64, prompt):
    """Make image summary"""
    chat = ChatOpenAI(model="gpt-4o-mini", max_tokens=1024)

    msg = chat.invoke(
        [
            HumanMessage(
                content=[
                    {"type": "text", "text": prompt},
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{img_base64}"},
                    },
                ]
            )
        ]
    )
    return msg.content


def generate_img_summaries(path):
    """
    Generate summaries and base64 encoded strings for images
    path: Path to list of .jpg files extracted by Unstructured
    """

    # Store base64 encoded images
    img_base64_list = []

    # Store image summaries
    image_summaries = []

    # Prompt
    prompt = """You are an assistant tasked with summarizing images for retrieval. \
    These summaries will be embedded and used to retrieve the raw image. \
    Give a concise summary of the image that is well optimized for retrieval."""

    # Apply to images
    for img_file in sorted(os.listdir(path)):
        if img_file.endswith(".jpg"):
            img_path = os.path.join(path, img_file)
            base64_image = encode_image(img_path)
            img_base64_list.append(base64_image)
            image_summaries.append(image_summarize(base64_image, prompt))

    return img_base64_list, image_summaries


# Image summaries
image_fpath = r'./figures'
img_base64_list, image_summaries = generate_img_summaries(image_fpath)
```

# 3. store those elements embeddings into the database

Store the raw texts, tables, and images in the docstore.

Store the texts, table summaries, and image summaries in the vectorstore for efficient semantic retrieval.

https://python.langchain.com/v0.1/docs/modules/data_connection/retrievers/multi_vector/#summary

```python
# Set environment variable to disable oneDNN optimizations to avoid kernel crash
# import os
# os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
```

```python
import uuid
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryStore
from langchain_chroma import Chroma
from langchain_core.documents import Document
from langchain_openai import OpenAIEmbeddings
```

```python
def create_multi_vector_retriever(
    vectorstore, text_summaries, texts, table_summaries, tables, image_summaries, images
):
    """
    Create retriever that indexes summaries, but returns raw images or texts
    """

    # Initialize the storage layer
    store = InMemoryStore()
    id_key = "doc_id"

    # Create the multi-vector retriever
    retriever = MultiVectorRetriever(
        vectorstore=vectorstore,
        docstore=store,
        id_key=id_key,
    )

    # Helper function to add documents to the vectorstore and docstore
    def add_documents(retriever, doc_summaries, doc_contents, doc_type):
        doc_ids = [str(uuid.uuid4()) for _ in doc_contents]
        documents = [
            Document(page_content=f"{s}\n\n{doc_type.upper()} CONTENT:\n{doc_contents[i]}", metadata={id_key: doc_ids[i], 'type': doc_type})
            for i, s in enumerate(doc_summaries)
        ]
        retriever.vectorstore.add_documents(documents)
        retriever.docstore.mset(list(zip(doc_ids, documents)))

    # Add texts, tables, and images
    if text_summaries:
        add_documents(retriever, text_summaries, texts, 'text')
    if table_summaries:
        add_documents(retriever, table_summaries, tables, 'table')
    if image_summaries:
        add_documents(retriever, image_summaries, images, 'image')

    return retriever

# Function to count the number of elements in the retriever
def count_elements_in_retriever(retriever):
    # Assuming the vectorstore has a method to get the number of documents
    vectorstore_count = retriever.vectorstore.count_documents()  # Replace with the actual method for your vectorstore
    docstore_count = len(retriever.docstore)  # InMemoryStore should support len()
    print(f"Number of elements in vectorstore: {vectorstore_count}")
    print(f"Number of elements in docstore: {docstore_count}")
    return vectorstore_count, docstore_count

# Function to retrieve and display results
def retrieve_and_display_results(retriever, query):
    try:
        results = retriever.invoke(query)
    except AttributeError:
        raise NotImplementedError("The retriever does not have an 'invoke' method. Please check the documentation for the correct query method.")
    
    # Loop through the results and display them based on metadata
    for result in results:
        if isinstance(result, Document):
            content = result.page_content
            metadata = result.metadata
            doc_type = metadata.get('type', 'text')
            display_result(content, doc_type)
        else:
            print("Retrieved result is not a Document object")

def is_image(content):
    try:
        image_data = base64.b64decode(content)
        image = Image.open(BytesIO(image_data))
        return True
    except Exception:
        return False

def display_result(content, doc_type):
    if doc_type == 'image' and is_image(content):
        image_data = base64.b64decode(content)
        image = Image.open(BytesIO(image_data))
        plt.imshow(image)
        plt.axis('off')
        plt.show()
    else:
        print(content)

```

```python
# The vectorstore to use to index the summaries
vectorstore3 = Chroma(
    collection_name="RAG_multi", embedding_function=OpenAIEmbeddings()
)

retriever_t2 = create_multi_vector_retriever(vectorstore3,
    text_summaries,
    texts_400_token,
    table_summaries,
    tables,
    image_summaries,
    img_base64_list,)
```

## build rag retriever

```python
# test
import base64
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from langchain_core.runnables import RunnablePassthrough

# Function to process documents and generate the context
def process_documents(documents):
    context = ""
    raw_docs = []
    for doc in documents:
        doc_type = doc.metadata.get('type')
        if doc_type == 'text':
            context += f"{doc.page_content}\n"
            raw_docs.append({'type': 'text', 'content': doc.page_content})
        elif doc_type == 'table':
            summary = doc.page_content.split('TABLE CONTENT:')[0]
            context += f"{summary}\n"
            raw_docs.append({'type': 'table', 'content': doc.page_content})
        elif doc_type == 'image':
            summary = doc.page_content.split('IMAGE CONTENT:')[0]
            context += f"{summary}\n"
            raw_docs.append({'type': 'image', 'content': doc.page_content})
    return context, raw_docs

# Function to display the raw documents
def display_raw_docs(raw_docs):
    for doc in raw_docs:
        if doc['type'] == 'text':
            print(f"Raw Document (Text):\n{doc['content']}\n")
        elif doc['type'] == 'table':
            print(f"Raw Document (Table):\n{doc['content']}\n")
        elif doc['type'] == 'image':
            image_data = doc['content'].split('IMAGE CONTENT:')[-1]
            image_data = base64.b64decode(image_data)
            image = Image.open(BytesIO(image_data))
            plt.imshow(image)
            plt.axis('off')
            plt.show()

def display_raw_docs_t(raw_docs):
    for i, doc in enumerate(raw_docs, start=1):
        print('---------------------------------------------')
        if doc['type'] == 'text':
            content = doc['content'].split('TEXT CONTENT:', 1)[0]
            print(f"Document {i} (Text):\n{content}\n")
            
        elif doc['type'] == 'table':
            print(f"Document {i} (Table):\n{doc['content']}\n")
        elif doc['type'] == 'image':
            print(f"Document {i} (Image):")
            image_data = doc['content'].split('IMAGE CONTENT:')[-1]
            image_data = base64.b64decode(image_data)
            image = Image.open(BytesIO(image_data))
            plt.imshow(image)
            plt.axis('off')
            plt.show()



# Define the prompt template
template = """Answer the question based only on the following context, which can include text and summary of tables or images:
{context}
Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)

# Define the model
model = ChatOpenAI(temperature=0.1, model="gpt-4o-mini")

# Function to handle the full pipeline
def retrieve_and_answer_question(retriever, query):
    # Retrieve documents based on the query
    documents = retriever.invoke(query)
    
    # Process the documents to create context and collect raw content
    context, raw_docs = process_documents(documents)
    
    # Prepare the prompt
    prompt_input = prompt.format(context=context, question=query)
    
    # Generate the response from the LLM
    llm_output = model.invoke(prompt_input)

    print('AI Chatbot:',llm_output.content)
    print('----------------------------------')
    
    # Display the raw documents
    display_raw_docs_t(raw_docs)

    
    # Return the final output
    return llm_output.content

```
